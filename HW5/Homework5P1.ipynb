```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading In the Images

trainingImage = cv2.imread('family.jpg')
trainingGroundTruth = cv2.imread('family.png', cv2.IMREAD_GRAYSCALE)

testingImage = cv2.imread('portrait.jpg')
testingGroundTruth = cv2.imread('portrait.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Naive Bayes Implementation

skinPixels = trainingImage[trainingGroundTruth == 255]
backgroundPixels = trainingImage[trainingGroundTruth == 0]

skinMean = np.mean(skinPixels, axis=0)
skinVariance = np.var(skinPixels, axis=0)
backgroundMean = np.mean(backgroundPixels, axis=0)
backgroundVariance = np.var(backgroundPixels, axis=0)

height, width, _ = testingImage.shape
classification = np.zeros((height, width), dtype=np.uint8)

for i in range(height):
    for j in range(width):
        pixel = testingImage[i, j, :2]
        skinProbability = (np.exp(-((pixel - skinMean[:2]) ** 2) / (2 * skinVariance[:2])) / np.sqrt(2 * np.pi * skinVariance[:2])).prod()
        backgroundProbability = (np.exp(-((pixel - backgroundMean[:2]) ** 2) / (2 * backgroundVariance[:2])) / np.sqrt(2 * np.pi * backgroundVariance[:2])).prod()
        classification[i, j] = 255 if skinProbability > backgroundProbability else 0
```

```python
# Displaying Test Results

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(testingImage, cv2.COLOR_BGR2RGB))
plt.title("Test Image")
plt.subplot(1, 3, 2)
plt.imshow(testingGroundTruth, cmap='gray')
plt.title("Ground Truth Mask")
plt.subplot(1, 3, 3)
plt.imshow(classification, cmap='gray')
plt.title("Classification Result")
plt.show()
```

```python
# Evaluation Metrics

TP = np.sum((classification == 255) & (testingGroundTruth == 255)) / np.sum(testingGroundTruth == 255)
TN = np.sum((classification == 0) & (testingGroundTruth == 0)) / np.sum(testingGroundTruth == 0)
FP = np.sum((classification == 255) & (testingGroundTruth == 0)) / np.sum(testingGroundTruth == 0)
FN = np.sum((classification == 0) & (testingGroundTruth == 255)) / np.sum(testingGroundTruth == 255)

print(f"True Positive: {round(TP * 100, 2)}%")
print(f"True Negative: {round(TN * 100, 2)}%")
print(f"False Positive: {round(FP * 100, 2)}%")
print(f"False Negative: {round(FN * 100, 2)}%")
```
